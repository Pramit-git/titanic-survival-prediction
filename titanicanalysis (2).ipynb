# **Titanic dataset**

The goal is to train a classifier that can predict the Survived column based on the other columns.

The dataset used here is from  https://homl.info/titanic.tgz

```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

def load_titanic_data():
    tarball_path = Path("datasets/titanic.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/titanic.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as titanic_tarball:
            titanic_tarball.extractall(path="datasets")
    return [pd.read_csv(Path("datasets/titanic") / filename)
            for filename in ("train.csv", "test.csv")]
```

The dataset is loaded. Let's split it up into training and test dataset.

```python
train_data, test_data = load_titanic_data()
```

```python
train_data.head()
```

In this dataset we can see that the passenger id column is like a unique number for each of the passengers.

So, the passenger id can be set as the index for both the datasets.

```python
train_data = train_data.set_index("PassengerId")
test_data = test_data.set_index("PassengerId")
```

We shall proceed with building our pipelines.

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])
```

Now for our categories

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
categ_pipeline = Pipeline([
        ("ordinal_encoder", OrdinalEncoder()),    
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("cat_encoder", OneHotEncoder(sparse_output=False)),
    ])
```

Now for joining the numerical and categorical pipelines together

```python
from sklearn.compose import ColumnTransformer

num_attribs = ["Age","SibSp","Parch","Fare"]
categ_attribs = ["Pclass","Sex","Embarked"]

preprocess_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", categ_pipeline, categ_attribs),
    ])
```

```python
X_train = preprocess_pipeline.fit_transform(train_data)
X_train
```

```python
y_train = train_data["Survived"]
```

We have selected "Survived" column as y because this is the column we have to predict.

Let's start with a random forest classifier. Because as per i know it is a lot advanced model.

```python
from sklearn.ensemble import RandomForestClassifier
forest_class = RandomForestClassifier(n_estimators=100, max_depth = 5, random_state=42)
forest_class.fit(X_train, y_train)
```

Now let's make predictions on test set

```python
X_test = preprocess_pipeline.transform(test_data)
y_pred = forest_class.predict(X_test)
```

Now let's check whether the model which we have built is working good or not. By performing cross validation

```python
from sklearn.model_selection import cross_val_score
forest_scores = cross_val_score(forest_class, X_train, y_train, cv=10)
forest_scores.mean()
```

Not bad. 82.38% accuracy. We should try more models.

Let's use a Support vector machine now

```python
from sklearn.svm import SVC

svm_class = SVC(gamma="auto")
svm_scores = cross_val_score(svm_class, X_train, y_train, cv=10)
svm_scores.mean()
```

A little bit higher. And this model looks better.

Let's use a gradient booster on the random forest model. It should make it better

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

gb_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
rf_pred = forest_class.predict(X_train)
print("Random Forest Accuracy:", accuracy_score(y_train, rf_pred))
gb_pred = gb_model.predict(X_train)
print("Gradient Boosting Accuracy:", accuracy_score(y_train, gb_pred))
```

Here we are! A 90.34% accuracy using the Gradient boost. That's the best model we have.
